# Lag Plots

(Sourced from: https://www.machinelearningplus.com/time-series/time-series-analysis-python/)

A Lag plot is a scatter plot of a time series against a lag of itself. It is normally used to check for autocorrelation. If there is any pattern existing in the series like the one you see below, the series is autocorrelated. If there is no such pattern, the series is likely to be random white noise.

In the example on Sunspots area time series, the plots get more and more scattered as the n_lag increases.

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
```

```python
# load data
url = "https://raw.githubusercontent.com/selva86/datasets/master/sunspotarea.csv"
sunspots_df = pd.read_csv(url, parse_dates=["date"], index_col="date")
url2 = "https://raw.githubusercontent.com/selva86/datasets/master/a10.csv"
a10_df = pd.read_csv(url2, parse_dates=["date"], index_col="date")
```

```python
# plot lags
fig, axes = plt.subplots(1, 4, figsize=(16,5), sharex=True, sharey=True)

for i, ax in enumerate(axes.flatten()[:4]):
    lag_plot(sunspots_df.value, lag=i+1, ax=ax, c='firebrick')
    ax.set_title('Lag ' + str(i+1))

fig.suptitle('Lag Plots of Sun Spots Area \n(Points get wide and scattered with increasing lag -> lesser correlation)\n', y=1.15)    
plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16,5), sharex=True, sharey=True)
for i, ax in enumerate(axes.flatten()[:4]):
    lag_plot(a10_df.value, lag=i+1, ax=ax, c='firebrick')
    ax.set_title('Lag ' + str(i+1))

fig.suptitle('Lag Plots of Drug Sales', y=1.05)    
plt.show()
```
