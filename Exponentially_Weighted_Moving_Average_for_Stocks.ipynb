```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
```

## Load stock data

```python
ticker = "AAPL"
aapl_df = yf.download(tickers=ticker, period="max", interval="1d", group_by="ticker", auto_adjust=True)
aapl_df
```

```python
aapl_df = aapl_df.loc["2019-01-01":]
aapl_df
```

## Visualise adjusted close prices

```python
plt.figure(figsize=(16,8))
plt.title("Adj Close Price History since 2017", fontsize=18)
plt.plot(aapl_df["Close"])
plt.xlabel("Date", fontsize=18)
plt.ylabel("Adj Close", fontsize=18)
plt.show()
```

## Exponentially Weighted Moving Average (EWMA)

```python
def EWMA(data, period, column="Close"):
  return data[column].ewm(span=period, adjust=False).mean()
```

```python
aapl_df.loc[:,"EWMA21"] = EWMA(aapl_df, 21)
```

```python
aapl_df.loc[:,"EWMA63"] = EWMA(aapl_df, 63)
```

```python
aapl_df.head(10)
```

## Visualise EMA on top of adjusted close prices

```python
aapl_df.loc[:,("Close", "EWMA21", "EWMA63")].plot(figsize=(16,8), legend=True, linewidth=2)
plt.legend(fontsize=18)
plt.title("Adj Close Price History since 2017 w/ EWMAs", fontsize=18)
plt.xlabel("Date", fontsize=18)
plt.ylabel("Adj Close", fontsize=18)
plt.show()
```
