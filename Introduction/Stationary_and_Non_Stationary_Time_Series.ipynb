- autocorrelation is the correlation of a time series with its previous values
- stationary time series are free from seasonal effects

## How to make a time series stationary (options)

- diiferencing the series (once or more until approx. stationary - most common)
- taking the log of the series
- taking the n-th root of the series
- a combination of the above listed options

## Differencing

- Differencing of a time series in discrete time is the transformation of the series to a new series where the values are the difference b/w consecutive values of the original time series.
- This procedure may be applied consecutively more than once, giving rise to the "first differences", "second differences", etc.

First differences in general: $d^{(1)} (t)=x(t)-x(t-1)$

Second differences in general: $d^{(2)} (t)=d^{(1)} (t)-d^{(1)} (t-1)$

m-th differences in general: $d^{(m)} (t)=d^{(m-1)} (t)-d^{(m-1)} (t-1)$

E.g. `time_series = [1, 2, 4, 7, 11]`
- `first_diff = [2-1, 4-2, 7-4, 11-7] = [1, 2, 3, 4]`
- `second_diff = [2-1, 3-2, 4-3] = [1, 1, 1]`

The differencing procedure, combined with the ARMA model for stationary time series, gives rise to the ARIMA model for non-stationary time series.


## Why make a non-stationary series stationary before forecasting?

- Forecasting stationary series is relatively easy and the forecasts are more reliable.
- Autoregressive forecasting models are essentially linear regression models that utilize the lag(s) of the series itself as predictors.
- Linear regression works best when the predictors (independent variables) are not correlated against each other.
- Stationarizing a series solves this problem since it removes any persistent autocorrelation, which makes the predictors (lags in the series) in the forecasting model nearly independent.

## Testing for stationarity

- eye inspection (like at the top of this page)
- split series into 2 or more contigious parts and compute summary statistics (mean, variance, autocorrelation, etc.). If the statistics are quite different, the series is not likely to be stationary.
- conduct "Unit Root Tests" to quantitatively determine whether series is stationary or not

## Unit Root Tests

There are multiple variations of this, where the tests check if a time series is non-stationary and possesses a unit root.

- Augmented Dickey-Fuller test (ADF Test) to check if time series has a unit root and is non-stationary
- Kwiatowski-Phillips-Schmidt-Shin test (KPSS test) for trend stationarity
- Philips-Perron test (PP test)
