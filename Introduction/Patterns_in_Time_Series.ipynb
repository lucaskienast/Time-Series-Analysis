# Patterns in Time Series

(Sourced from: https://www.machinelearningplus.com/time-series/time-series-analysis-python/)

Additive Time Series = Base Lvl + Trend + Seasonality + Error

Multiplicative Time Series = Base Lvl * Trend * Seasonality * Error

- trend: increasing or decreasing slope over time
- seasonaility: distinct repeated pattern between regular intervals due to seasonal factors
- not all time series must have trends or seasonality (might just have one or none)
- cyclic: distinct repeated patterns between regular intervals due to socio-economic factors (not seasonal)

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.style.use("seaborn-whitegrid")
```

## Load data

```python
trend_df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/guinearice.csv", parse_dates=["date"], index_col="date")
seasonality_df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/sunspotarea.csv", parse_dates=["date"], index_col="date")
trend_and_seasonality_df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/AirPassengers.csv", parse_dates=["date"], index_col="date")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(24,6))
trend_df.plot(title="Trend", legend="False", ax=axes[0])
seasonality_df.plot(title="Seasonality", legend="False", ax=axes[1])
trend_and_seasonality_df.plot(title="Trend & Seasonality", legend="False", ax=axes[2])
plt.show()
```
